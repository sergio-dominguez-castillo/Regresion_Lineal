3. Construye un modelo de regresión lineal que relacione el volumen y el peso de los
peces. Realiza una breve evaluación del modelo y grafícalo.

```python
#Definir las variables predictoras (X) y la variable de respuesta (y)
X = df_fish["Volumen"]
y = df_fish["Weight"]

# Agregar una constante al predictor (intercepto)
X = sm.add_constant(X)

# Crear el modelo de regresión lineal
model = sm.OLS(y, X).fit()

# Obtener los resultados del modelo
results = model.summary()

# Imprimir los resultados
print(results)

# Agregar una constante al predictor (intercepto)
X = sm.add_constant(X)
# Obtener las predicciones del modelo
prediccion= model.predict(X)

# Calcular el Error Cuadrático Medio (MSE) para cada modelo
mse = ((df_fish['Weight'] - prediccion) ** 2).mean()
# Calcular el Error Cuadrático Medio Explicado (EMSE) para cada modelo
emse = mse / ((df_fish['Weight'] - df_fish['Weight'].mean() ) ** 2).mean()
# Calcular el Coeficiente de Determinación (R²) para cada modelo
r2 = model.rsquared
# Calcular el Error Absoluto Medio (MAE) para cada modelo
mae = (abs(df_fish['Weight'] - prediccion)).mean() 

# Mostrar los resultados
print("---------------------------------------------------")
print("Modelo prediccion peso de peces:")
print(f"R²: {r2:.2f}, MSE: {mse:.2f}, EMSE: {emse:.2f}, MAE: {mae:.2f}")
print("---------------------------------------------------")

# Graficar la regresión lineal
plt.scatter(df_fish["Volumen"], df_fish["Weight"], label="Datos")
plt.plot(df_fish["Volumen"], model.predict(X), color='red', label="Regresión Lineal")
plt.title("Regresión lineal entre el volumen y el peso de los peces")
plt.xlabel("Volumen")
plt.ylabel("Peso de los peces")
plt.legend()
plt.show()
```

Nota:
al analizar el MAE: 66.03, existe error en la prediccion que corresponde a 66,03 (onzas) 


4. ¿Qué sucede si aplicas el modelo anterior por separado a cada especie de peces?
Explica.

```python
# especies
df_fish["Species"].unique()
```

```python
# Ciclo para recorre las especies de peces
# llamar a funcion para graficar
for peces in df_fish["Species"].unique():
    #Definir las variables predictoras (X)
    X = df_fish.loc[df_fish["Species"]==peces,["Volumen"]]
    # Agregar una constante al predictor (intercepto)
    X = sm.add_constant(X)
    # Obtener las predicciones del modelo
    prediccion= model.predict(X)
    # datos reales
    real= df_fish.loc[df_fish["Species"]==peces,["Volumen","Weight"]]

    # Calcular el Error Cuadrático Medio (MSE) para cada modelo
    mse = ((real['Weight'] - prediccion) ** 2).mean()
    # Calcular el Error Cuadrático Medio Explicado (EMSE) para cada modelo
    emse = mse / ((real['Weight'] - real['Weight'].mean() ) ** 2).mean()
    # Calcular el Coeficiente de Determinación (R²) para cada modelo
    r2 = model.rsquared
    # Calcular el Error Absoluto Medio (MAE) para cada modelo
    mae = (abs(real['Weight'] - prediccion)).mean() 
    print("------------------------------------------------")
    # Mostrar los resultados
    print(f"Modelo para pez de la especie -> {peces}")
    print(f"R²: {r2:.2f}, MSE: {mse:.2f}, EMSE: {emse:.2f}, MAE: {mae:.2f}")
    print("------------------------------------------------")
    #graficar
    fn.Graficar_Regresion(real['Volumen'],real['Weight'],prediccion,"Volumen","Peso",peces)
```

Nota:
Las especies que tiene menor MAE, sus valores reales estan mas cercanos a la regresion lineal, efectivamente tiene una alta tasa de efectividad, respecto a las otras especies que tienen valores mas alejados.
podria deducir que el modelo al ser utilizado para precedir los pesos d elos peces separados por especies va a tener un comportamiento para predecir mejor (sera mas preciso) para aquellas especies que sus valores reales estan mas cerca a la regrecion lineal.

5. Construye un modelo de regresión que relacione dos o más variables con el peso de
los peces. Realiza una breve evaluación del modelo.

```python
#Definir las variables predictoras (X) y la variable de respuesta (y)
X = df_fish[["Length1","Length2","Length3"]]
y = df_fish["Weight"]

# Agregar una constante al predictor (intercepto)
X = sm.add_constant(X)

# Crear el modelo de regresión lineal
model = sm.OLS(y, X).fit()

# Obtener los resultados del modelo
results = model.summary()

# Imprimir los resultados
print(results)
```

```python
prediccion= model.predict(X)

# Calcular el Error Cuadrático Medio (MSE) para cada modelo
mse = ((df_fish['Weight'] - prediccion) ** 2).mean()
# Calcular el Error Cuadrático Medio Explicado (EMSE) para cada modelo
emse = mse / ((df_fish['Weight'] - df_fish['Weight'].mean() ) ** 2).mean()
# Calcular el Coeficiente de Determinación (R²) para cada modelo
r2 = model.rsquared
# Calcular el Error Absoluto Medio (MAE) para cada modelo
mae = (abs(df_fish['Weight'] - prediccion)).mean() 

# Mostrar los resultados
print("-------------------------------------------")
print("Modelo prediccion peso de peces:")
print(f"R²: {r2:.2f}, MSE: {mse:.2f}, EMSE: {emse:.2f}, MAE: {mae:.2f}")
print("-------------------------------------------")

```

```python
# Graficar la regresión lineal
plt.scatter(df_fish["Length1"], df_fish["Weight"], label="Datos length1", color='blue', alpha=0.5)
plt.scatter(df_fish["Length2"], df_fish["Weight"], label="Datos length2", color='yellow', alpha=0.5)
plt.scatter(df_fish["Length3"], df_fish["Weight"], label="Datos length3", color='green', alpha=0.5)

plt.plot(df_fish["Length1"], model.predict(X), color='blue', label="Regresión Lineal length1")
plt.plot(df_fish["Length2"], model.predict(X), color='yellow', label="Regresión Lineal length2")
plt.plot(df_fish["Length3"], model.predict(X), color='green', label="Regresión Lineal length3")

plt.title("Regresión lineal entre el longitud 1,2,3  y el peso de los peces")
plt.xlabel("Longitud 1,2,3")
plt.ylabel("Peso de los peces")
plt.legend()
plt.show()
```
